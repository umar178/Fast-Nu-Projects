Question # 01:

```python
import time
import random

class SmartStreetLight:
    def __init__(self):
        self.light_intensity = None
        self.motion_detected = None
        self.status = "LIGHT OFF"

    def read_light_sensor(self):
        self.light_intensity = random.choice(["Bright", "Dim"])
    
    def read_motion_sensor(self):
        self.motion_detected = random.choice(["Yes", "No"])

    def decide_action(self):
        if self.light_intensity == "Bright":
            self.status = "LIGHT OFF"
        elif self.light_intensity == "Dim" and self.motion_detected == "Yes":
            self.status = "LIGHT ON"
        else:
            self.status = "LIGHT OFF"

    def apply_action(self):
        print(f"Light Intensity: {self.light_intensity}")
        print(f"Motion Detected: {self.motion_detected}")
        print(f"Street Light Status: {self.status}")
        print("-" * 40)

    def run(self):
        while True:
            self.read_light_sensor()
            self.read_motion_sensor()
            self.decide_action()
            self.apply_action()
            time.sleep(1)

street_light_agent = SmartStreetLight()
street_light_agent.run()
```

Question # 02:

```python
class FloorCleaningRobot:
    def __init__(self):
        self.rooms = {
            1: "Dirty",
            2: "Dirty",
            3: "Dirty"
        }
        self.position = 1

    def perceive(self):
        return self.position, self.rooms[self.position]

    def update_state(self, room, status):
        self.rooms[room] = status

    def all_clean(self):
        return all(status == "Clean" for status in self.rooms.values())

    def decide_action(self, position, status):
        if status == "Dirty":
            return "Clean"

        if position == 1:
            return "Move Right"
        elif position == 2:
            if self.rooms[3] == "Dirty":
                return "Move Right"
            elif self.rooms[1] == "Dirty":
                return "Move Left"
        elif position == 3:
            return "Move Left"
        return None

    def act(self, action):
        if action == "Clean":
            self.rooms[self.position] = "Clean"
            print(f"Room {self.position} cleaned.")
        elif action == "Move Right":
            self.position += 1
            print(f"Moving right to Room {self.position}.")
        elif action == "Move Left":
            self.position -= 1
            print(f"Moving left to Room {self.position}.")

    def run(self):
        step = 1
        while not self.all_clean():
            pos, status = self.perceive()
            print(f"Step {step}: In Room {pos}, Status: {status}")
            action = self.decide_action(pos, status)
            if not action:
                break
            self.act(action)
            step += 1
        print("\nAll rooms are clean!")
        print("Final States:", self.rooms)


robot = FloorCleaningRobot()
robot.run()
```

Question # 03:

```python
class WarehouseEnvironment:
    def __init__(self):
        self.goal = 15

    def is_goal_reached(self, position):
        return position == self.goal


class DeliveryRobotAgent:
    def __init__(self):
        self.position = 0
        self.visited_positions = [self.position]
        self.steps = 0

    def decide_action(self):
        if self.position < 15:
            return +1
        elif self.position > 15:
            return -1
        return 0

    def act(self, action):
        self.position += action
        self.visited_positions.append(self.position)
        self.steps += 1
        return self.position


env = WarehouseEnvironment()
agent = DeliveryRobotAgent()

while not env.is_goal_reached(agent.position):
    action = agent.decide_action()
    agent.act(action)

print("Positions visited:", agent.visited_positions)
print("Total steps taken:", agent.steps)
```

```python
class DeliveryOption:
    def __init__(self, name, time, cost):
        self.name = name
        self.time = time
        self.cost = cost

    def compute_utility(self, weight_time, weight_cost):
        return weight_time * (1 / self.time) + weight_cost * (1 / self.cost)


class ShoppingAssistantAgent:
    def __init__(self, weight_time, weight_cost):
        self.weight_time = weight_time
        self.weight_cost = weight_cost

    def recommend(self, optionA, optionB):
        utilA = optionA.compute_utility(self.weight_time, self.weight_cost)
        utilB = optionB.compute_utility(self.weight_time, self.weight_cost)

        if utilA > utilB:
            return optionA.name, utilA
        else:
            return optionB.name, utilB


optionA = DeliveryOption("Option A", time=1, cost=50)
optionB = DeliveryOption("Option B", time=3, cost=20)

agent = ShoppingAssistantAgent(weight_time=0.6, weight_cost=0.4)
recommended_option, score = agent.recommend(optionA, optionB)
print(f"Recommended: {recommended_option}  Utility Score: {score:.3f}")
```
